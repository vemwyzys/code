```python
import torch

# 创建一个张量x，并指明需要计算梯度
x = torch.tensor([2.0], requires_grad=True)

# 定义一个操作
y = x**2 + 3 * x + 5

# 查看x的梯度
print(x.grad)

# 对y进行反向传播，计算梯度
y.backward()
print('y', y)

# 查看x的梯度
print(x.grad)  # 输出：tensor([11.])，因为y对x的导数是2*x + 3，当x=2时导数值是7
```

```python
import numpy as np

# 假设的权重和偏置
w = 0.5
b = -300

# 输入特征（房屋面积）
x = np.array([800, 1000, 1200])

# 线性回归预测
y_pred = w * x + b
print("Linear Regression Output:", y_pred)
```

```python
def relu(x):
    return np.maximum(0, x)

# 使用ReLU激活函数
y_pred_relu = relu(w * x + b)
print("Output with ReLU Activation:", y_pred_relu)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 生成随机数据
np.random.seed(0)
x1 = 2.5 * np.random.randn(100) + 1.5   # 特征1
x2 = 1.5 * np.random.randn(100) + 0.5   # 特征2
res = 0.5 * np.random.randn(100)       # 生成残差
y = 2 + 0.3 * x1 + 0.5 * x2 + res      # 目标变量

# 创建一个 3D 图形对象
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制数据点
ax.scatter(x1, x2, y, color='b', marker='o', label='Actual data')

# 创建网格以计算平面
x1_grid, x2_grid = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
y_grid = 2 + 0.3 * x1_grid + 0.5 * x2_grid  # 平面方程

# 绘制拟合平面
ax.plot_surface(x1_grid, x2_grid, y_grid, color='r', alpha=0.5, linewidth=0)

ax.set_xlabel('Feature 1 ($x_1$)')
ax.set_ylabel('Feature 2 ($x_2$)')
ax.set_zlabel('Target ($y$)')
ax.legend()

# 显示图形
plt.show()
```

```python
import numpy as np

# 定义矩阵 P
P = np.array([[0.4, 0.5],
              [0.6, 0.5]])

# 定义向量 v
v = np.array([0.4, 0.6])

# 第一次乘法
result1 = np.dot(P, v)
print("第一次乘法结果:", result1)

result2 = np.dot(P, result1)
print("第二次乘法结果:", result2)



result3 = np.dot(P, result2)
print("第三次乘法结果:", result3)

result4 = np.dot(P, result3)
print("第4次乘法结果:", result4)

result5 = np.dot(P, result4)
print("第5次乘法结果:", result5)

result6 = np.dot(P, result5)
print("第6次乘法结果:", result6)

result7 = np.dot(P, result6)
print("第7次乘法结果:", result7)
```

```python
import numpy as np

# 定义矩阵 P
P = np.array([[0.4, 0.5],
              [0.6, 0.5]])

# 定义初始向量 v
v = np.array([0, 1])

# 执行100次矩阵与向量的乘法
for i in range(100):
    v = np.dot(P, v)
    print(f"迭代 {i+1} 的结果: {v}")
```
