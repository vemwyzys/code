```python
import math
for i in range(-10, 11):
    value = math.exp(i)
    print(f"e^{i}={value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个从-2到2（包含）的x值数组，共100个点
x = np.linspace(-2, 2, 100)
# 计算e的x次方
y = np.exp(x)

# 创建图形
plt.figure(figsize=(8, 4))
# 绘制曲线
plt.plot(x, y, label='$e^x$')
# 添加图例
plt.legend()
# 添加标题
plt.title('Curve of $e^x$')
# 添加x轴和y轴标签
plt.xlabel('x')
plt.ylabel('$e^x$')
# 显示网格（可选）
plt.grid(True)
# 显示图形
plt.show()
```
