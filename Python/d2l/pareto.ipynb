```python
import numpy as np
import matplotlib.pyplot as plt

# 生成随机数据
np.random.seed(0)
data = np.random.rand(100, 2) * 100

# 排序数据（根据第一个目标）
sorted_data = data[data[:,0].argsort()]

# 找到帕累托边界
pareto_front = []
current_y = float('inf')
for x, y in sorted_data:
    if y < current_y:
        pareto_front.append((x, y))
        current_y = y

pareto_front = np.array(pareto_front)

# 绘图
plt.scatter(data[:, 0], data[:, 1], label='Data Points')
plt.plot(pareto_front[:, 0], pareto_front[:, 1], color='red', label='Pareto Front')
plt.title('Pareto Boundary Example')
plt.xlabel('Objective 1')
plt.ylabel('Objective 2')
plt.legend()
plt.grid(True)
plt.show()
```
