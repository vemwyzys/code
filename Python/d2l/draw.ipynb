```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 用于3D图

# 生成示例数据
np.random.seed(42)  # 为了重现性
X = np.random.rand(100, 2)  # 生成100个样本，每个样本有2个特征
weights = np.array([200000, -5000])  # 假设的真实权重，第一个特征（面积）的权重是200000，第二个特征（房龄）的权重是-5000
intercept = 50000  # 截距项

# 计算目标值y
y = X @ weights + intercept + np.random.randn(100) * 10000  # 加入一些噪音

plt.scatter(X[:, 0], y)  # 以第一个特征（面积）和目标值（价格）之间的关系为例
plt.title('House Price vs. Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()

```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 用散点图表示数据点
ax.scatter(X[:, 0], X[:, 1], y, c='b', marker='o')

ax.set_xlabel('Area')
ax.set_ylabel('Age')
ax.set_zlabel('Price')
plt.title('House Price by Area and Age')
plt.show()
```

```python
# 创建网格
area = np.linspace(X[:, 0].min(), X[:, 0].max(), 10)
age = np.linspace(X[:, 1].min(), X[:, 1].max(), 10)
area, age = np.meshgrid(area, age)
Z = area * weights[0] + age * weights[1] + intercept

# 绘制平面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(area, age, Z, alpha=0.5, rstride=100, cstride=100, color='r', edgecolor='none')
ax.scatter(X[:, 0], X[:, 1], y, c='b', marker='o')

ax.set_xlabel('Area')
ax.set_ylabel('Age')
ax.set_zlabel('Price')
plt.title('House Price Prediction by Linear Regression Model')
plt.show()
```
