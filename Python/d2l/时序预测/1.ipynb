```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 创建示例时间序列数据
np.random.seed(42)
time = np.arange(100)
data = 0.5 * time + 10 * np.sin(time) + np.random.normal(size=100)

# 将数据转换为DataFrame
df = pd.DataFrame({'time': time, 'data': data})

# 绘制原始时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['data'], label='Original Time Series')
plt.title('Original Time Series')
plt.legend()
plt.show()

# 进行一次差分(即计算当前值与前一个值的差。)
df['diff_data'] = df['data'].diff()

# 绘制差分后的时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['diff_data'], label='First Difference')
plt.title('First Difference of Time Series')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 创建示例时间序列数据
np.random.seed(42)
time = np.arange(200)
seasonal_period = 12
data = 10 * np.sin(2 * np.pi * time / seasonal_period) + np.random.normal(size=200)

# 将数据转换为DataFrame
df = pd.DataFrame({'time': time, 'data': data})

# 绘制原始时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['data'], label='Original Time Series')
plt.title('Original Time Series')
plt.legend()
plt.show()

# 进行季节性差分
df['seasonal_diff_data'] = df['data'].diff(seasonal_period)

# 绘制季节性差分后的时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['seasonal_diff_data'], label='Seasonal Difference')
plt.title('Seasonal Difference of Time Series')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 创建示例时间序列数据（含趋势和季节性成分）
np.random.seed(42)
time = np.arange(200)
trend = 0.5 * time
seasonal_period = 12
seasonal = 10 * np.sin(2 * np.pi * time / seasonal_period)
noise = np.random.normal(size=200)
data = trend + seasonal + noise

# 将数据转换为DataFrame
df = pd.DataFrame({'time': time, 'data': data})

# 绘制原始时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['data'], label='Original Time Series')
plt.title('Original Time Series')
plt.legend()
plt.show()

# 进行一次差分以消除趋势
df['first_diff'] = df['data'].diff()

# 绘制一次差分后的时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['first_diff'], label='First Difference')
plt.title('First Difference of Time Series')
plt.legend()
plt.show()

# 进行季节性差分以消除季节性成分
df['seasonal_diff'] = df['data'].diff(seasonal_period)

# 绘制季节性差分后的时间序列
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['seasonal_diff'], label='Seasonal Difference')
plt.title('Seasonal Difference of Time Series')
plt.legend()
plt.show()
```
