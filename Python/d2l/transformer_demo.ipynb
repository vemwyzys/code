```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 创建一个新的3D图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 点的坐标
# 假设'男生'在(1, 1, 1)，'女生'在(2, 2, 2)，这样它们之间的距离较近
# '变形金刚'在(5, 5, 5)，距离稍远
# '黑洞'在(10, 10, 10)，距离最远
points = {
    'girl': (1, 1, 1),
    'boy': (4, 2, 2),
    'Transformers': (5, 5, 15),
    'black hole': (20, 20, 20)
}

# 为每个点添加标记，并在旁边标注名字
for point, coords in points.items():
    ax.scatter(*coords)  # 绘制点
    ax.text(coords[0], coords[1], coords[2], '%s' % (point), size=10, zorder=2, color='k')  # 添加文本注释

# 设置图例
ax.legend()
# 指定图像大小
ax.set_xlim([0, 25])
ax.set_ylim([0, 25])
ax.set_zlim([0, 25])

# 设置坐标轴标签
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# 展示图形
plt.show()
```
