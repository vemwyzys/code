```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
# ''(1, 1, 1)''(2, 2, 2)
# ''(5, 5, 5)
# ''(10, 10, 10)
points = {
    'girl': (1, 1, 1),
    'boy': (4, 2, 2),
    'Transformers': (5, 5, 15),
    'black hole': (20, 20, 20)
}

# 
for point, coords in points.items():
    ax.scatter(*coords)  # 
    ax.text(coords[0], coords[1], coords[2], '%s' % (point), size=10, zorder=2, color='k')  # 

# 
ax.legend()
# 
ax.set_xlim([0, 25])
ax.set_ylim([0, 25])
ax.set_zlim([0, 25])

# 
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# 
plt.show()
```
